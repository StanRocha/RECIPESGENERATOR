# Separate the dataset by classes and download the images for the correct folder Train&Test

# Chicken

```python
#Create a dataset with just the information of chicken category
Chicken=new_recipes[new_recipes['RecipeCategory']=='Chicken'].reset_index(drop=True)
```

```python
dict_link=dict(zip(Chicken['RecipeId'],Chicken['image_link'] ))
# dict_link
```

```python
#Download 2000 photos to train

bad_list=[]
limit=0

for i, url2 in list(dict_link.items())[:]:
    counter = 0
    for u in url2:
        limit+=1
        counter = counter + 1
        identifier = str(i) + "_____" + str(counter)
        if limit <= 2000:
        
            try:
                urllib.request.urlretrieve(u, "C:\\Users\\User\\Desktop\\CLEAN_FINAL_PROJECT\\images\\Train\\Chicken\\chicken" + str(identifier) + ".jpg")


            except:
                bad_list.append(str(identifier))
                print(bad_list)
                pass
        else:
            break
        
```

```python
#Download 500 photos to test

bad_list=[]
limit=0

for i, url2 in list(dict_link.items())[-2500:]:
    counter = 0
    for u in url2:
        limit+=1
        counter = counter + 1
        identifier = str(i) + "_____" + str(counter)
        if limit <= 500:
        
            try:
                urllib.request.urlretrieve(u, "C:\\Users\\User\\Desktop\\CLEAN_FINAL_PROJECT\\images\\Test\\Chicken\\chicken" + str(identifier) + ".jpg")


            except:
                bad_list.append(str(identifier))
                print(bad_list)
                pass
        else:
            break
        
```
