
Cheking the 'RecipeIngredientQuantities' and 'RecipeIngredientParts' there's no meaning to be separate so i will agregate the two columns


```python
# after taking a look in the columns i understand that they have similarities so i will do a variable that can clean both columns

def clean_str(row):
  clean_list=[]
  try:
    clean_list.append(row.replace('c("',' ').replace('"',"").strip(')').strip())
    return clean_list
  except:
    return "TAKE A LOOK"
```

```python
#use the apply mode to change all the rows and create new columns
new_recipes["clean_ingredient"] = new_recipes['RecipeIngredientQuantities'].apply(clean_str)
```

```python
new_recipes["clean_recipe"] = new_recipes["RecipeIngredientParts"].apply(clean_str)
```

```python
new_recipes.head(2)
```

```python
#drop the old columns
new_recipes=new_recipes.drop(["RecipeIngredientQuantities","RecipeIngredientParts"], axis=1)
```

```python
new_recipes["clean_ingredient"][0]
#i need separate the quantities and the ingredientes so i will use apply to split
new_recipes["clean_ingredient2"] = new_recipes["clean_ingredient"].apply(lambda x: x[0].split(','))
new_recipes["clean_recipe2"] = new_recipes["clean_recipe"].apply(lambda x: x[0].split(','))
```

```python
def zipped(row):
  a = row["clean_ingredient2"]
  b = row["clean_recipe2"]
  return list(zip(a,b))

new_recipes["ingredients"] = new_recipes.apply(zipped, axis=1)
new_recipes["ingredients"]
```

```python
#change the type to clean characters from 'ingredients', using replace leaving just que quantities and the ingredients
new_recipes['ingredients']=new_recipes['ingredients'].astype(str)
```

```python
# i will do a for loop to split the links in each row

split_link=[]

for row in new_recipes['Images']:
  split_link.append(row.split('"'))

new_recipes['split_link']=split_link

new_recipes['split_link'][0]
```

```python
# I did a variable to append just the link and then use the apply to use in all rows

def find_jpg(array):
    abc =[]
    for el in array:
        if el.endswith(".jpg"):
            abc.append(el)
        elif el.endswith(".JPG"):
            abc.append(el)
        elif el.endswith(".jpeg"):
            abc.append(el)
        else:
            pass
    return abc
```

```python
new_recipes['image_link']=new_recipes['split_link'].apply(lambda x : find_jpg(x))
new_recipes['image_link'][0]
```
